```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Unorganisedfile.csv")
```

```python
df
```

**now we can see that Columns "name", "Age" and etc have missing values.**

```python
df.columns
```

```python
#handling missing values
df[' Name'] = df[' Name'].replace(' ', 'UNKNOWN')

df[' Age']=pd.to_numeric(df[' Age'], errors='coerce')
median_age=df[' Age'].median()
df[' Age'].fillna(median_age, inplace=True)

df[' Join Date']=df[' Join Date'].replace(' ', '2021-01-12')
df
```

```python
#handling formatting as we see Income column have also non-numeric values
df[' Income']=pd.to_numeric(df[' Income'], errors='coerce')
```

**We see that the someone salary is in negative which is not possible and some salary is exceptionally high, that are outliers. lets see how we handle them through pandas**

```python
#handling outliers 
#using machine learning concept (IQR) and box-plot way
q1 = df[' Income'].quantile(0.25) #quantile is used to calculate percentile in pandas
q3 = df[' Income'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df[' Income'] = pd.to_numeric(df[' Income'], errors='coerce')
df = df[(df[' Income'] >= lower_bound) & (df[' Income'] <= upper_bound)]
df
```

<h2> This is the proper cleaned data without missing values, outliers and irregular formattings</h2>
