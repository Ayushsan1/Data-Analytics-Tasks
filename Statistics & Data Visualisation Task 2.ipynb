```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv('pre_cleaned_dataset1.csv')
```

```python
df.columns
```

```python
#statistics Calculations
mean_age=df['Age'].mean()
mode_age=df['Age'].mode()[0]
median_age=df['Age'].median()
std_age=df['Age'].std()

print("Mean Age of Data is : ", mean_age)
print("Maximum Occuring Age is: ", mode_age)
print("Median(avg) age of data is: ", median_age)
print("Standard Devaition of Age from data is: ", std_age)
```

# Visualisation Methods #

```python
#histogram 
plt.figure(figsize=(10,5))
sns.histplot(df['Age'], bins=10, kde=True)
#kde is a kernel density estimation and represents PDF(probab. density function)
plt.title('Histogram Plot- Age vs Income')
```

<h1><span style="color:red">Conclusion Analysis:</span></h1> 
<h3>This particular histogram, the income appears to be higher for younger age groups (around 30 years old) and then lowers for older age groups (around 50 years old). Itâ€™s important to note that because this is a histogram, we can only say that income is higher for younger age groups on average. It is possible that there are some older people who make more money than some younger people </h3>

```python
#scatterPlot 
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Age', y='Income', data=df)
plt.title('Age vs. Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

<h1><span style="color:red">Conclusion Analysis:</span></h1> 
<h3> The scatter plot shows a positive correlation between age and income. This means that as age increases, 
income tends to increase as well. However, the correlation is not perfect. There are many factors that can affect 
a persons income, such as education, experience, and occupation.</h3>   

```python
#lineplot of same graph
sns.lineplot(df, x='Age', y='Income', label="lineplot visualisation")
```

**In this graph we can easily undertsand the trend of income with age**
